```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import matplotlib.pyplot as plt

theta_a = []
#for posterior distribution
mu_a = [0.5, 0]
sigma_a = [[0.3, 0], [0, 2]]

x_a = []
y_a = []

theta_b = []
#for posterior distribution
mu_b = [0, 0.5]
sigma_b = [[2, 0], [0, 0.3]]

x_b = []
y_b = []

np.random.seed(0)
theta_a = np.random.multivariate_normal(mu_a, sigma_a, 1000)
theta_b = np.random.multivariate_normal(mu_b, sigma_b, 1000)
plt.scatter( theta_a[:,0], theta_a[:,1], color='red')
plt.scatter( theta_b[:,0], theta_b[:,1], color='blue')
plt.show
```

```python
np.random.seed(0)
x_a = np.zeros((1000,2))
x_a[:,0] = np.random.uniform(0,1, 1000)

x_b =  np.zeros((1000,2))
x_b[:,1] = np.random.uniform(0,1, 1000)


plt.scatter( x_a[:,0], x_a[:,1], color='red')
plt.scatter( x_b[:,0], x_b[:,1], color='blue')
plt.show
```

```python
mu_e = [0.5, 0.5]
sigma_e = [[2, 0.3], [2, 0.3]]
epsilon = np.random.multivariate_normal(mu_e, sigma_e, 1000)

y_a = x_a * theta_a + epsilon
y_b = x_b * theta_b + epsilon

plt.scatter( y_a[:,0], y_a[:,1], color='red')
plt.scatter( y_b[:,0], y_b[:,1], color='blue')
plt.show
```
